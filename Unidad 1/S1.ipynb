# Comprobar datos

```python
column = 'semestre'

plt.figure(figsize=(10, 6))
plt.hist(df[column], bins=range(1, 12), edgecolor='black', align='left', rwidth=0.8)

# Customize the x-ticks to match the numbers in the data
plt.xticks(range(min(df[column]), max(df[column])+1 ))

# Show the plot
plt.title(f'Histogram of {column}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
column = 'promedio'

# Define los bordes de los bins para crear exactamente 10 barras entre 60 y 100
bins = np.linspace(60, 100, 11)  # 10 intervalos -> 11 bordes

plt.figure(figsize=(10, 6))
plt.hist(df[column], bins=bins, edgecolor='black', align='mid', rwidth=0.8)

# Ajusta las marcas del eje X para que coincidan con los valores medios de los bins
plt.xticks(np.arange(60, 101, 4))  # Aquí 4 es el ancho de cada intervalo (100-60)/10

# Mostrar el gráfico con título y etiquetas
plt.title(f'Histogram of {column}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
column = 'ingresos'

plt.figure(figsize=(10, 6))
plt.hist(df[column], bins=np.arange(int(min(df[column])), int(max(df[column])) + 2, max(df[column] - min(df[column]))/10), 
         edgecolor='black', align='left', rwidth=0.8)

# Customize the x-ticks to match the numbers in the data
plt.xticks(np.arange(min(df[column]), max(df[column])+1, 5000))

# Show the plot
plt.title(f'Histogram of {column}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
column = 'opinion_campus'

plt.figure(figsize=(10, 6))
plt.hist(df[column], bins=range(1, 12), edgecolor='black', align='left', rwidth=0.8)

# Customize the x-ticks to match the numbers in the data
plt.xticks(range(min(df[column]), max(df[column])+1 ))

# Show the plot
plt.title(f'Histogram of {column}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
column = 'trabaja'

plt.figure(figsize=(10, 6))
plt.hist(df[column].astype(int), 
         bins=[-0.5, 0.5, 1.5],  # Definimos bordes para alinear las barras con 0 y 1
         edgecolor='black', align='mid', rwidth=0.8)

# Configurar las etiquetas del eje X como 'False' y 'True'
plt.xticks([0, 1], labels=['False', 'True'])

# Títulos y etiquetas
plt.title(f'Histogram of {column}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
column = 'distancia_universidad'

plt.figure(figsize=(10, 6))
plt.hist(df[column], 
         bins=10, 
         edgecolor='black', align='left', rwidth=0.8)
plt.title(f'Histogram of {column}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
column = 'edad'

plt.figure(figsize=(10, 6))
plt.hist(df[column], 
         bins=np.arange(int(min(df[column])), int(max(df[column])) + 2, max(df[column] - min(df[column]))/10), 
         edgecolor='black', align='left', rwidth=0.8)
plt.title(f'Histogram of {column}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
column = 'tipo_vivienda'

plt.figure(figsize=(10, 6))
plt.hist(df[column], 
         bins=2, 
         edgecolor='black', align='mid', rwidth=0.8)
plt.title(f'Histogram of {column}')
plt.xlabel('Value')
plt.xticks(df[column].unique())
plt.ylabel('Frequency')
plt.show()
```
