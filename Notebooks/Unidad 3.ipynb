# Cargar librerías

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
```

# Generar datos

```python
# Parameters
mean = 5
std = 1
num_data = 100000


# Generate data
data = np.random.normal(mean, std, num_data)
data = pd.DataFrame({'data': data})
data
```

# Jugar con el histograma para ver los datos de muestras aleatorias

```python
#sample = data.sample(1000, random_state=42)
sample = data.sample(1000)
num_bins = 10

plt.figure(figsize=(10, 6))
plt.hist(sample, bins=num_bins, density=False, alpha=0.6, color='b')
plt.title(f'Histograma')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
plt.show()
```

# Jugar con las curvas

```python
mean = sample['data'].mean()
std = sample['data'].std()
min = sample['data'].min()
max = sample['data'].max()

print(f'La media de la muestra es: {mean}')
print(f'La desviación estándar de la muestra es: {std}')
print(f'El valor mínimo de la muestra es: {min}')
print(f'El valor máximo de la muestra es: {max}')
```

```python
x = np.linspace(min, max, 100)
p = norm.pdf(x, mean, std)

plt.figure(figsize=(10, 6))
plt.plot(x, p, 'k', linewidth=2, label='Curva aproximada')
plt.title('Curva Gaussiana con los valores aproximados')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

# Graficar ambos

```python
plt.figure(figsize=(10, 6))

plt.hist(sample, bins=num_bins, density=True, alpha=0.6, color='b')
plt.plot(x, p, 'k', linewidth=2, label='Curva aproximada')

plt.title('Histograma y su respectiva curva')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

# Ver curvas con parámetros aproximados vs. curva real

## Curva real

```python
mean = data['data'].mean()
std = data['data'].std()
min = data['data'].min()
max = data['data'].max()

print(f'La media real: {mean}')
print(f'La desviación estándar real: {std}')
print(f'El valor mínimo real: {min}')
print(f'El valor máximo real: {max}')

x = np.linspace(min, max, 100)
p = norm.pdf(x, mean, std)

plt.figure(figsize=(10, 6))
plt.plot(x, p, 'k', linewidth=2, label='Curva real')
plt.title('Curva Gaussiana con los valores aproximados')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

## Curva aproximada

```python
sample_x = data.sample(1000)

mean = sample_x['data'].mean()
std = sample_x['data'].std()
min = sample_x['data'].min()
max = sample_x['data'].max()

print(f'La media de la muestra es: {mean}')
print(f'La desviación estándar de la muestra es: {std}')
print(f'El valor mínimo de la muestra es: {min}')
print(f'El valor máximo de la muestra es: {max}')

x = np.linspace(min, max, 100)
p = norm.pdf(x, mean, std)

plt.figure(figsize=(10, 6))
plt.plot(x, p, linewidth=2, c='b', label='Curva aproximada')
plt.title('Curva Gaussiana con los valores aproximados')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

## Curva real vs. varias aproximaciones

```python
mean = data['data'].mean()
std = data['data'].std()
min = data['data'].min()
max = data['data'].max()


sample_1 = data.sample(1000)
mean_1 = sample_1['data'].mean()
std_1 = sample_1['data'].std()

sample_2 = data.sample(1000)
mean_2 = sample_2['data'].mean()
std_2 = sample_2['data'].std()

sample_3 = data.sample(100)
mean_3 = sample_3['data'].mean()
std_3 = sample_3['data'].std()

sample_4 = data.sample(10)
mean_4 = sample_4['data'].mean()
std_4 = sample_4['data'].std()


print('----------------------------------------')
print('COLOR NEGRO: VALORES REALES')
print(f'La media real: {mean}')
print(f'La desviación estándar real: {std}')
print(f'El valor mínimo real: {min}')
print(f'El valor máximo real: {max}')
print('----------------------------------------')


print('----------------------------------------')
print('COLOR AZUL: MUESTRA 1 (1000 DATOS)')
print(f'La media de la muestra es: {mean_1}')
print(f'La desviación estándar de la muestra es: {std_1}')
print('----------------------------------------')


print('----------------------------------------')
print('COLOR NARANJA: MUESTRA 2 (1000 DATOS)')
print(f'La media de la muestra es: {mean_2}')
print(f'La desviación estándar de la muestra es: {std_2}')
print('----------------------------------------')


print('----------------------------------------')
print('COLOR ROJO: MUESTRA 3 (100 DATOS)')
print(f'La media de la muestra es: {mean_3}')
print(f'La desviación estándar de la muestra es: {std_3}')
print('----------------------------------------')

print('----------------------------------------')
print('COLOR ROJO: MUESTRA 4 (10 DATOS)')
print(f'La media de la muestra es: {mean_4}')
print(f'La desviación estándar de la muestra es: {std_4}')
print('----------------------------------------')

x = np.linspace(min, max, 100)
p = norm.pdf(x, mean, std)
p_1 = norm.pdf(x, mean_1, std_1)
p_2 = norm.pdf(x, mean_2, std_2)
p_3 = norm.pdf(x, mean_3, std_3)
p_4 = norm.pdf(x, mean_4, std_4)

plt.figure(figsize=(10, 6))
plt.plot(x, p, color='black', linewidth=2, label='Curva con datos reales')
plt.plot(x, p_1, color='blue', linewidth=2, label='Curva con muestra 1 (1000 datos)')
plt.plot(x, p_2, color='orange', linewidth=2, label='Curva con muestra 2 (1000 datos)')
plt.plot(x, p_3, color='red', linewidth=2, label='Curva con muestra 3 (100 datos)')
plt.plot(x, p_4, color='cyan', linewidth=2, label='Curva con muestra 4 (10 datos)')
plt.title('Distintas Curvas Gaussiana con los valores aproximados y los reales')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```
